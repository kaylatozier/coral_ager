```python
import csv
import argparse
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import detrend
from scipy.signal import find_peaks
from scipy.interpolate import interp1d
```

```python
df_d18o = pd.read_csv("simulated_d18o_dataset.csv")  # columns: depth_mm, delta18O
```

```python
df_sst = pd.read_csv("simulated_sst_dataset.csv")    # columns: years_ago, sst (deg C)
```

```python
# Detrend both series and add columns to the datasets
df_d18o["detrended"] = detrend(df_d18o["d18o (per mil)"])
df_sst["detrended"] = detrend(df_sst["SST (°C)"])
```

```python
# z-score normalization so that signals can be easily compared; mean = 0, std = 1
df_sst["normalized"] = (df_sst["SST (°C)"] - df_sst["SST (°C)"].mean()) / df_sst["SST (°C)"].std()
df_d18o["normalized"] = (df_d18o["d18o (per mil)"] - df_d18o["d18o (per mil)"].mean()) / df_d18o["d18o (per mil)"].std()

# Summer!
# finding sst peaks 
sst_summer_peaks, _ = find_peaks(df_sst["normalized"], distance=6)  # corals grow roughly 6 mm per season, so wait at least 6 peaks before picking the highest one
# Find δ18O troughs (we negate the normalized δ18O to find troughs instead of peaks)
d18o_summer_troughs, _ = find_peaks(-df_d18o["normalized"], distance=6)

# Winter!

# finding sst troughs
sst_winter_troughs, _ = find_peaks(-df_sst["normalized"], distance=6)
# finding d18o peaks
d18o_winter_peaks, _ = find_peaks(df_d18o["normalized"], distance=6)
```

```python
# concatenate two arrays
total_sst_peaks = np.concatenate((sst_summer_peaks, sst_winter_troughs))
total_d18o_peaks = np.concatenate((d18o_summer_troughs, d18o_winter_peaks))
```

```python
# sort arrays
total_sst_peaks.sort()
total_d18o_peaks.sort()
```

```python
# extract peaks from datasets
sst_anchors = df_sst.iloc[total_sst_peaks] # .iloc returns only the rows at the trough locations
d18o_anchors = df_d18o.iloc[total_d18o_peaks] # .iloc returns only the rows at the trough locations
```

```python
# creating new dataset with the age points
num_tiepoints = min(len(total_sst_peaks), len(total_d18o_peaks)) # limit the matching to the shorter list to avoid index errors (if there are more sst peaks than d18o troughs, for example)
anchor_depths = d18o_anchors["Depth (mm)"].values[:num_tiepoints] # new array of depths with only the numebr of tiepoints
anchor_ages = sst_anchors["Years Ago"].values[:num_tiepoints] # new array of ages with only the number of tiepoints
```

```python
print(num_tiepoints)

print(anchor_depths)

print(anchor_ages)
```

```python
# Plot δ¹⁸O vs. Depth with Selected Tie Points
plt.figure(figsize=(10, 6))

# Plot δ¹⁸O vs Depth 
plt.plot(df_d18o["Depth (mm)"], df_d18o["d18o (per mil)"], label="δ¹⁸O vs. Depth", color='b')

# Plot the selected tie points 
plt.scatter(d18o_anchors["Depth (mm)"], d18o_anchors["d18o (per mil)"], color='r', label='Selected Tie Points', zorder=5)

# Adding labels and title
plt.xlabel('Depth (mm)', fontsize=12)
plt.ylabel('δ¹⁸O (per mil)', fontsize=12)
plt.title('δ¹⁸O vs. Depth with Selected Tie Points', fontsize=14)
# invert axes
plt.gca().invert_xaxis() 
# Display grid and legend
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

```python
# Plot δ¹⁸O vs. Depth with Selected Tie Points
plt.figure(figsize=(10, 6))

# Plot δ¹⁸O vs Depth 
plt.plot(df_sst["Years Ago"], df_sst["SST (°C)"], label="SST vs. Year", color='b')

# Plot the selected tie points 
plt.scatter(sst_anchors["Years Ago"], sst_anchors["SST (°C)"], color='r', label='Selected Tie Points', zorder=5)

# Adding labels and title
plt.xlabel('Years Ago', fontsize=12)
plt.ylabel('SST (°C)', fontsize=12)
plt.title('SST vs. Years Ago with Selected Tie Points', fontsize=14)

# Display grid and legend
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```

```python
# interpolate the full age model
interp_func = interp1d(anchor_depths, anchor_ages, kind='linear', fill_value="extrapolate") # fill_value = extrapolate: points outside the data range will be extrapolated
df_d18o["age_model"] = interp_func(df_d18o["Depth (mm)"]) # create a new column with the age model
```

```python
print(df_d18o["age_model"])
```

```python
# Interpolate δ¹⁸O over even time steps
even_time = np.arange(df_d18o["age_model"].min(), df_d18o["age_model"].max(), 0.1)
even_d18o = np.interp(even_time, df_d18o["age_model"], df_d18o["d18o (per mil)"])
```

```python
plt.plot(df_d18o["age_model"], df_d18o["d18o (per mil)"], label="δ¹⁸O vs. Age Model", color='g')
plt.xlabel('Age (Years Ago)', fontsize=12)
plt.ylabel('δ¹⁸O (per mil)', fontsize=12)
plt.title('Age Model vs. δ¹⁸O', fontsize=14)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.grid(True)
plt.legend()
plt.show()
```
